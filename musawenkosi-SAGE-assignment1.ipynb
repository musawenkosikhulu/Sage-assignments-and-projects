(*) Write a recursive function with the name `my_fibonacci_number` for the Fibonacci recurrence equation: 

$$F_{n} = F_{n-1} + F_{n-2}, \qquad F_{0} = 1, \quad F_{1} = 1.$$

Your function should return the $n$-th [Fibonacci number](https://en.wikipedia.org/wiki/Fibonacci_number).

```python
def my_fibonacci_number(n):
    """
    This Fibonacci sequence function takes a positive integer argument and
    return an n-th term of the Fibonacci number
    """
    if n <= 1:     #This conditional statement checks if the number n enter is less or equals to 1 then returns 0
        return 0
    elif n ==1:    #In this case if n is 1 the function will return 1 because we do not have f_-1
        return 2
    else:          #This code creates fibonacci start from 1 as specified in the question 
        F_0 = 1
        F_1 = 1
        for i in range(2,n+1): #This code generates a loop  for recussive nature of fibonacci
            F_n = F_0 + F_1    #This updates the value of F_n for every loop
            F_0 = F_1          #This updates F_1 from the previous value
            F_1 = F_n          #This updates F_1 from the previous value
        return F_n             #This returns the n-th term of the Fibonacci number
    
my_fibonacci_number(4)         #This tests out if the code works
```

What do you observe about the movement of the robot when `number_of_walks` takes on the values: $100, 1000, 5000$.

```python
def random_walk_on_graph(my_graph,number_of_walks): #This was with the help of Tebogo who is one of the January intakes
    Dictionary=my_graph.to_dictionary()
    Keys=list(Dictionary.keys()) # Convert the dictionary keys into a list
    start=choice(Keys)      # name will be picked randomly
    List=[start]
    for i in range(number_of_walks):
        y=choice(Dictionary[start])
        if y in Keys:
            List.append(y)
        start=y
        
    Destination_of_robot=[]
    Number_of_visits=[]
    for k in set(List):
        c=List.count(k)
        Destination_of_robot.append(k)
        Number_of_visits.append(c)
    W=table([Destination_of_robot,Number_of_visits],header_column=['Verices','Number of visits'])
    return W
    pass
```
