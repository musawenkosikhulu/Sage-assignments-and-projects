For sake of example we will use only 100 demonstrate. The code for the example is given below. 

```python
def random_walk_on_graph(my_graph,number_of_walks):
    Dictionary=my_graph.to_dictionary()
    Keys=list(Dictionary.keys()) # Convert the dictionary keys into a list
    start=choice(Keys)      # name will be picked randomly
    List=[start]
    for i in range(number_of_walks):
        y=choice(Dictionary[start])
        if y in Keys:
            List.append(y)
        start=y
        
    Destination_of_robot=[]
    Number_of_visits=[]
    for k in set(List):
        c=List.count(k)
        Destination_of_robot.append(k)
        Number_of_visits.append(c)
    W=table([Destination_of_robot,Number_of_visits],header_column=['Verices','Number of visits'])
    return W
    pass
```
